```python
import time
from collections import deque
import copy

# BFS_maze function
def bfs_maze(maze, start, goal):
    rows, cols = len(maze), len(maze[0])
    visited = set()
    parent = {}
    def is_valid(x, y):
        return 0 <= x < rows and 0 <= y < cols and maze[x][y] != '#' and (x, y) not in visited
    queue = deque([start])
    visited.add(start)
    found = False
    while queue:
        x, y = queue.popleft()
        if (x, y) == goal:
            found = True
            break
        moves = [(0, 1), (1, 0), (0, -1), (-1, 0)]
        for dx, dy in moves:
            new_x, new_y = x + dx, y + dy
            if is_valid(new_x, new_y):
                queue.append((new_x, new_y))
                visited.add((new_x, new_y))
                parent[(new_x, new_y)] = (x, y)
    if not found:
        return "No path found", 0
    path = []
    x, y = goal
    while (x, y) != start:
        path.append((x, y))
        x, y = parent[(x, y)]
    path.append(start)
    path.reverse()
    for x, y in path:
        if maze[x][y] != 'S' and maze[x][y] != 'G':
            maze[x][y] = '*'
    return "Path found", len(path)

# DFS_maze function
def dfs_maze(maze, start, goal):
    rows, cols = len(maze), len(maze[0])
    visited = set()
    path = []
    def is_valid(x, y):
        return 0 <= x < rows and 0 <= y < cols and maze[x][y] != '#' and (x, y) not in visited
    def dfs(x, y):
        if (x, y) == goal:
            path.append((x, y))
            return True
        visited.add((x, y))
        path.append((x, y))
        moves = [(0, 1), (1, 0), (0, -1), (-1, 0)]
        for dx, dy in moves:
            new_x, new_y = x + dx, y + dy
            if is_valid(new_x, new_y):
                if dfs(new_x, new_y):
                    return True
        path.pop()
        return False
    if dfs(start[0], start[1]):
        for x, y in path:
            if maze[x][y] != 'S' and maze[x][y] != 'G':
                maze[x][y] = '*'
        return "Path found", len(path)
    else:
        return "No path found", 0

```

```python
#  BFS 
def bfs_maze(maze, start, goal):
    rows, cols = len(maze), len(maze[0])
    visited = set()
    parent = {}
    queue = deque([start])
    visited.add(start)

    while queue:
        x, y = queue.popleft()
        if (x, y) == goal:
            break
        for dx, dy in [(0,1),(1,0),(0,-1),(-1,0)]:
            nx, ny = x+dx, y+dy
            if 0<=nx<rows and 0<=ny<cols and maze[nx][ny]!='#' and (nx,ny) not in visited:
                queue.append((nx,ny))
                visited.add((nx,ny))
                parent[(nx,ny)] = (x,y)

    steps = 0
    if goal in parent or start==goal:
        x, y = goal
        steps = 1
        while (x, y) != start:
            x, y = parent[(x,y)]
            steps += 1
    return steps

# DFS 
def dfs_maze(maze, start, goal):
    rows, cols = len(maze), len(maze[0])
    visited = set()
    steps = [0]

    def dfs(x, y, count):
        if (x,y)==goal:
            steps[0] = count+1
            return True
        visited.add((x,y))
        for dx, dy in [(0,1),(1,0),(0,-1),(-1,0)]:
            nx, ny = x+dx, y+dy
            if 0<=nx<rows and 0<=ny<cols and maze[nx][ny]!='#' and (nx,ny) not in visited:
                if dfs(nx, ny, count+1):
                    return True
        return False

    dfs(start[0], start[1], 0)
    return steps[0]

maze = [
    ['S', '.', '.', '.', '.', '.', '#', '.', '.'],
    ['#', '#', '#', '#', '#', '.', '#', '#', '#'],
    ['.', '.', '.', '.', '.', '.', '.', '.', 'G'],
    ['#', '#', '#', '#', '#', '#', '#', '#', '#']
]

#  Test different start/end points 
points = [((0,0),(2,8)), ((2,0),(0,8)), ((0,1),(2,6)), ((0,3),(2,3))]

for start, goal in points:
    bfs_steps = bfs_maze(copy.deepcopy(maze), start, goal)
    dfs_steps = dfs_maze(copy.deepcopy(maze), start, goal)
    print(f"Start: {start}, Goal: {goal}")
    print(f"BFS Steps: {bfs_steps}")
    print(f"DFS Steps: {dfs_steps}")
    print(" ")
```
