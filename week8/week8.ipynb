running the codes

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import make_blobs
from sklearn.cluster import KMeans
from sklearn.preprocessing import StandardScaler


#  Spherical clusters (K-Means works)
data_spherical, _ = make_blobs(n_samples=300, centers=3, cluster_std=1.0, random_state=42)

kmeans = KMeans(n_clusters=3, random_state=42)
labels = kmeans.fit_predict(data_spherical)

plt.scatter(data_spherical[:,0], data_spherical[:,1], c=labels, cmap="viridis")
plt.scatter(kmeans.cluster_centers_[:,0], kmeans.cluster_centers_[:,1], 
            c="red", s=200, marker="x")
plt.title("K-Means Works on Spherical Clusters")
plt.show()
```

```python
#  Non-spherical clusters (K-Means fails)

np.random.seed(42)
cluster_1 = np.dot(np.random.randn(150,2), [[4,1],[1,2]]) + [5,10]
cluster_2 = np.dot(np.random.randn(150,2), [[4,-1],[-1,2]]) + [25,25]

data_nonspherical = np.vstack([cluster_1, cluster_2])

plt.scatter(data_nonspherical[:,0], data_nonspherical[:,1])
plt.title("Non-Spherical Data")
plt.show()

kmeans = KMeans(n_clusters=2, random_state=42)
labels = kmeans.fit_predict(data_nonspherical)

plt.scatter(data_nonspherical[:,0], data_nonspherical[:,1], c=labels, cmap="plasma")
plt.title("K-Means Fails on Non-Spherical Clusters")
plt.show()
```

```python

# Part 3: Different feature scales (K-Means fails)

group_1 = np.random.normal([0,0], [1,40], size=(150,2))
group_2 = np.random.normal([7,7], [40,1], size=(150,2))
data_unscaled = np.vstack([group_1, group_2])

kmeans = KMeans(n_clusters=2, random_state=42)
labels = kmeans.fit_predict(data_unscaled)

plt.scatter(data_unscaled[:,0], data_unscaled[:,1], c=labels, cmap="coolwarm")
plt.title("K-Means Without Scaling  Incorrect")
plt.show()
```

```python
# Part 4: After feature scaling (K-Means works)

scaler = StandardScaler()
data_scaled = scaler.fit_transform(data_unscaled)

kmeans = KMeans(n_clusters=2, random_state=42)
labels = kmeans.fit_predict(data_scaled)

plt.scatter(data_scaled[:,0], data_scaled[:,1], c=labels, cmap="coolwarm")
plt.title("K-Means With Scaling  Correct")
plt.show()
```

```python
#  Importing required libreraies

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

from sklearn.preprocessing import StandardScaler
from sklearn.cluster import KMeans
from sklearn.metrics import silhouette_score
from scipy.cluster.hierarchy import dendrogram, linkage
```

```python
#  load dataset
df = pd.read_csv("OnlineRetail.csv", encoding="ISO-8859-1")
df.head()

```

```python
#  overview dataset

df.info()
df.describe()
```

```python
# data cleaning

df=df.dropna(subset=['CustomerID'])
df=df[df["Quantity"] > 0]
df=df[df["UnitPrice"] > 0]
```

```python
df['TotalPrice'] = df['Quantity'] * df['UnitPrice']
```

```python
df['InvoiceDate'] = pd.to_datetime(df['InvoiceDate'], dayfirst=True)
```

```python
reference_date = df['InvoiceDate'].max()

rfm = df.groupby('CustomerID').agg({
    'InvoiceDate': lambda x: (reference_date - x.max()).days,
    'InvoiceNo': 'count',
    'TotalPrice': 'sum'
})

rfm.columns = ['Recency', 'Frequency', 'Monetary']
rfm.head()
```

```python
#  feature scaling
scaler = StandardScaler()
rfm_scaled = scaler.fit_transform(rfm)
```

```python
# Elbow Method

inertia = []

for k in range(2, 11):
    kmeans = KMeans(n_clusters=k, random_state=42)
    kmeans.fit(rfm_scaled)
    inertia.append(kmeans.inertia_)

plt.plot(range(2, 11), inertia, marker='o')
plt.xlabel('Number of Clusters')
plt.ylabel('Inertia')
plt.title('Elbow Method')
plt.show()
```

```python
#K-Means Clustering
def plot_kmeans_clusters(data, n_clusters):
    kmeans = KMeans(n_clusters=n_clusters, random_state=random_state_value)
    labels = kmeans.fit_predict(data)
    
    # Plot Recency vs Monetary
    plt.scatter(data[:,0], data[:,2], c=labels, cmap='viridis', edgecolor='k', s=50)
    plt.scatter(kmeans.cluster_centers_[:,0], kmeans.cluster_centers_[:,2], 
                c='red', marker='x', s=200, label='Centroids')
    plt.xlabel('Recency')
    plt.ylabel('Monetary')
    plt.title(f'K-Means Clustering with {n_clusters} Clusters')
    plt.legend()
    plt.show()
```

```python
from sklearn.cluster import KMeans


N = [2, 3, 4, 5, 6, 7]
for n in N:
    plot_kmeans_clusters(rfm_scaled, n_clusters=n)
```
