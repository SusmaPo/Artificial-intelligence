```python
# Classes
#!/usr/bin/python
class Employee: #name of class is Employee
    'Common base class for all employees'
    empCount = 0

    def __init__(self, name, salary):
        self.name=name
        self.salary= salary
        Employee.empCount +=1

    def displayCount(self): #further methods are defined like functions
        print("total Employee %d" % Employee.empCount)
    
    def displayEmployee(self):
        print("Name : ", self.name, ", Salary:", self.salary )
"This would create first object of Employee  class"
emp1=Employee("Zara", 2000)
"This would create second object of Employee  class"
emp2=Employee("Manni", 5000)
emp1.displayEmployee()
emp2.displayEmployee()
print ("Total Employee %d" % Employee.empCount)

# \n newlinw; \r carriage return; \t tab
```

Exercise 1

Create Python code to convert temprature between Fahrenheit and centigrade. the temprature conversion formulae are:

Centigrade= (Farenheit-32)/1.8

Farenheit=(Centigrade*1.8)+32


```python
# Conversion of frahentite to centigrade

fahrenheit=float(input("Enter temperature in fahrenheit: "))
centigrade=(fahrenheit-32)/1.8
print(f"{fahrenheit}F° is equal to {centigrade:.2f}°C")


centigrade=float(input("Enter temperature in centigrade: "))
fahrenheit=(centigrade*1.8)+32
print(f"{centigrade}C° is equal to {fahrenheit:.2f}°F")
```
