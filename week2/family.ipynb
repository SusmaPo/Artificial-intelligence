```python
import json
from kanren import run, facts,eq,Relation, var, conde
from kanren. constraints import neq
```

```python
father_dictionary={
    'John':['Ram'],
    'William':['Doe']
}
```

```python
mother_dictionary={
    'Megan':['William']
    
}
```

```python
def father(x,y):
    if x in father_dictionary and y in father_dictionary[x]:
        print(f"{x} is the parent of {y}")
    else:
        print(f'{x} is not the parent of {y}')
```

```python
def mother(x,y):
    if x in mother_dictionary and y in father_dictionary[x]:
        print(f"{x} is the parent of {y}")
    else:
        print(f'{x} is not the parent of {y}')
```

```python
def parent(x,y):
    return conde ([father(x,y)], [mother(x,y)])

def grandparent(x,y):
    temp = var()
    return conde ((parent(x,temp), parent(temp,y)))

def sibling(x,y):
    temp = var()
    return conde ((parent(temp,x), parent(temp,y), neq(x,y)))

def uncle(x,y):
    parent_temp = var()
    return conde(
        (parent(parent_temp, y), sibling(x, parent_temp), father(x,var()))
    )
```

```python
if __name__ =="__main__":
    father = Relation()
    mother= Relation()

    with open('relationships.json') as f:
        d=json.loads(f.read())

    for item in d['father']:
        facts(father,(list(item.keys())[0],list(item.values())[0]))
    
    for item in d['mother']:
        facts(mother,(list(item.keys())[0],list(item.values())[0]))
    
    x=var()
```

Question 3: write a script to find all spouses

```python
def spouse(x,y):
    c=var()
    return conde((father(x,c), mother(y,c)))

X,Y=var(),var()
spouses=run(0,(X,Y), spouse(X, Y))

for f, m in set(spouses):
    print(f"{f} is the spouse of {m}")
```
